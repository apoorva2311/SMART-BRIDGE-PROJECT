```python
#IMPORT LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy.stats as stats
```

```python
data=pd.read_csv('HumanBreastCancer_data_set.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data['diagnosis'].unique()
```

```python
X=data.drop(['id','diagnosis'],axis=1)
X
```

```python
def enc(a):
    if a=='M':
        return 1
    else:
        return 0
```

```python
y=data['diagnosis'].apply(enc)
```

```python
y.head()
```

```python
X.head()
```

```python
X.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scale=MinMaxScaler()
```

```python
a=scale.fit(X)
```

```python
x=a.transform(X)
```

```python
X=pd.DataFrame(x,columns=X.columns)
```

```python
X.head()
```

```python
col=X.columns
```

```python
for i in col:
    print(np.corrcoef(X[i],y))
```

```python
X.drop(columns=["area_mean","perimeter_mean","perimeter_worst","Unnamed: 32"],inplace=True)
```

```python
X.columns
```
